```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/ML Lab Assignments/Assignment 1/Heart.csv')
df.head()
```

```python
# find mean, median, mode

df.mean()
```

```python
df.median()
```

```python
df.mode()
```

```python
# find min value, max value and range

df.min()
```

```python
df.max()
```

```python
# Convert boolean columns to integers (0 and 1)
for col in df.select_dtypes(include=['bool']).columns:
    df[col] = df[col].astype(int)

# Now calculate the range
range_df = df.max() - df.min()
print(range_df)
```

```python
# find standard deviation, variance and percentile

df.std()
```

```python
# find variance

df.var()
```

```python
# prompt: find percentile of data

# find percentile
print(df.quantile(0.25)) # 25th percentile
print('-'*20)
print(df.quantile(0.50)) # median or 50th percentile
print('-'*20)
print(df.quantile(0.75)) # 75th percentile
print('-'*20)
```
